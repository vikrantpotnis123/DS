```python
majors = recent_grads['Major'].value_counts(normalize=True, dropna=True).index.tolist()
```

```python
major_median_pay = {}
majors = recent_grads['Major'].value_counts(normalize=True, dropna=True).index.tolist()

for m in majors:
    selected_rows = recent_grads.loc[recent_grads["Major"] == m]['Median']
    major_median_pay[m] = selected_rows.mean()
```

```python
major_total = {}
for m in majors:
    selected_rows = recent_grads.loc[recent_grads["Major"] == m]['Total']
    major_total[m] = selected_rows.mean()
```

```python
major_median_pay_series = pd.Series(major_median_pay).sort_values(ascending=False)
df1 = pd.DataFrame(major_median_pay_series, columns=['Major Median'])
print(df1)
```

```python
major_total_series = pd.Series(major_total).sort_values(ascending=False)
df2 = pd.DataFrame(major_total_series, columns=['Major Total'])
```
